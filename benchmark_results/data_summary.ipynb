```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# plt.rc('font', family='Arial')
```

```python
def plot(file_a, file_b, title):
    # Load data from CSV files
    df_a = pd.read_csv(file_a)
    df_b = pd.read_csv(file_b)
    
    # Task 1: Sort data
    df_a.sort_values(by='name of the test file', inplace=True)
    df_b.sort_values(by='filename', inplace=True)
    
    # Task 2: Extract columns
    solving_time_a = df_a['solving time sat']
    solving_time_b = df_b['solvingTime']
    
    # Task 3: Convert non-numeric values to 180 and convert to float
    solving_time_a = pd.to_numeric(solving_time_a, errors='coerce').fillna(180).astype(float)
    solving_time_b = pd.to_numeric(solving_time_b, errors='coerce').fillna(180).astype(float)

    num_greater_eq = np.sum(solving_time_b.values <= solving_time_a.values)
    print(f'AlloyMax is faster or equal to Flie in: {num_greater_eq}/{len(solving_time_a)}', num_greater_eq / len(solving_time_a))
    print(f'AlloyMax is faster than Flie by:', np.mean(solving_time_b), np.mean(solving_time_a), np.mean(solving_time_a)/np.mean(solving_time_b))
    
    
    # Set up the figure
    fig, ax = plt.subplots()
    
    # Task 4: Draw a scatter plot with zorder parameter
    sc = ax.scatter(solving_time_a, solving_time_b, zorder=3, alpha=0.7, s=100)
    
    # Task 5: Create x, y labels
    ax.set_xlabel('Flie', fontsize=24)
    ax.set_ylabel('ATLAS', fontsize=24)
    ax.tick_params(axis='both', which='major', labelsize=22)
    
    ax.set_xlim(-5, 185)
    ax.set_ylim(-5, 185)
    # Task 6: Draw a dashed line to the max value of the axis ticks
    ax.plot(ax.get_xlim(), ax.get_ylim(), '--', zorder=2, color='silver')
    # ax.plot((-5, 185), (-2.5, 92.5), 'r--', zorder=2, label="y=x/2")
    
    # Task 7: Ensure dots are above the axes and ticks
    ax.set_axisbelow(False)

    # ax.set_title(title)
    # ax.legend()
    
    # Show the plot
    plt.tight_layout()
    plt.savefig(f"{title.replace(' ', '_')}.png")
    plt.show()
```

```python
def load_results(file_a, file_b):
    # Load data from CSV files
    df_a = pd.read_csv(file_a)
    df_b = pd.read_csv(file_b)
    
    # Task 1: Sort data
    df_a.sort_values(by='name of the test file', inplace=True)
    df_b.sort_values(by='filename', inplace=True)
    
    # Task 2: Extract columns
    solving_time_a = df_a['solving time sat']
    solving_time_b = df_b['solvingTime']
    
    # Task 3: Convert non-numeric values to 180 and convert to float
    solving_time_a = pd.to_numeric(solving_time_a, errors='coerce').fillna(180).astype(float)
    solving_time_b = pd.to_numeric(solving_time_b, errors='coerce').fillna(180).astype(float)

    num_greater_eq = np.sum(solving_time_b.values <= solving_time_a.values)
    print(f'AlloyMax is faster or equal to Flie in: {num_greater_eq}/{len(solving_time_a)}', num_greater_eq / len(solving_time_a))
    print(f'AlloyMax is faster than Flie by:', np.mean(solving_time_b), np.mean(solving_time_a), np.mean(solving_time_a)/np.mean(solving_time_b))

    return solving_time_a, solving_time_b
```

```python
df_peterson_a, df_peterson_b = load_results('peterson/peterson_stats.csv', 'peterson/peterson_alloy.csv')
df_vote_a, df_vote_b = load_results('voting/voting_stats.csv', 'voting/voting_alloy.csv')
df_repair_a, df_repair_b = load_results('robot/robot_stats.csv', 'robot/robot_alloy.csv')
df_weakening_a, df_weakening_b = load_results('weakening/weakening_stats_exact_sol.csv', 'weakening/weakening_alloy_gr1_new.csv')
```

```python
def load_ltl_sketch(a):
    df = pd.read_csv(a, header=None)
    # df.columns = ['filename', 'expected', 'expected2', 'time', 'found']
    df.loc[df[df.columns[-1]] == False, df.columns[-2]] = 180.0

    print('Avarage solving time:', np.mean(df[df.columns[-2]]))
    
    return df[df.columns[-2]]
```
